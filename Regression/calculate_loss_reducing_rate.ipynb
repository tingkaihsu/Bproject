```python
import matplotlib.pyplot as plt
def CalcuLossRate(infilepath, nor_losses):
    with open(infilepath, 'r') as Input:
        lines = Input.readlines()
#     Output = open(outfilepath, 'w')
    losses = []
#     x = []
    i = 1
    for line in lines:
        loss = float(line)
        losses.append(loss)
#         x.append(i)
        i = i + 1
    maximum = losses[0]
#     nor_losses = []
    for loss in losses:
        nor_losses.append(loss/maximum)

            
```

```python
def Plot(log_cosh, mse):
    x = []
    for i in range(1, 71):
        x.append(i)
#         print(i)
    fig, ax = plt.subplots()
    ax.set_title("Validation_loss Reduction Rate")
    plt.xlabel("Epoches")
    plt.ylabel("Normalized Vlidation_loss")
    ax.scatter(x,log_cosh, linewidth=2.0, color = 'red')
    ax.scatter(x, mse, linewidth=2.0, color = 'blue')
    plt.legend(
    loc='best',
    labels = ['log_cosh', 'mse'])
#     plt.show()
    plt.savefig('val_loss.png')
```

```python
log_cosh = []
mse = []
CalcuLossRate('log_cosh.txt', log_cosh)
CalcuLossRate('MSE.txt', mse)
# print(log_cosh)
# print(mse)
Plot(log_cosh, mse)
```

```python
def PlotMass(filepath):
    x = []
    for i in range(1, 486):
        x.append(i)
        i += 1
    with open(filepath, 'r') as file:
        lines = file.readlines()
    masses = []
    for line in lines:
        mass = float(line)
        masses.append(mass)
        
        
    fig, ax = plt.subplots()
    ax.set_title("Mass Prediction")
    plt.xlabel("Number of Prediction")
    plt.ylabel("Mass")
    ax.scatter(x, masses, linewidth=2.0, color = 'blue')
#     plt.legend(
#     loc='best',
#     labels = ['log_cosh', 'mse'])
#     plt.show()
    plt.savefig('mass_predict.png')
```
