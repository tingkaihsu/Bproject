```python
import matplotlib.pyplot as plt
def CalcuLossRate(infilepath):
    with open(infilepath, 'r') as Input:
        lines = Input.readlines()
#     Output = open(outfilepath, 'w')
    losses = []
    x = []
    i = 1
    for line in lines:
        loss = float(line)
        losses.append(loss)
        x.append(i)
        i = i + 1
    maximum = losses[0]
    nor_losses = []
    for loss in losses:
        nor_losses.append(loss/maximum)
    fig, ax = plt.subplots()
    ax.set_title("Validation_loss Reduction Rate")
    plt.xlabel("Epoches")
    plt.ylabel("Vlidation_loss")
    ax.scatter(x,nor_losses, linewidth=2.0)
#     plt.show()
    plt.savefig('test.png')
            
```

```python
CalcuLossRate('loss_log.txt')
```
